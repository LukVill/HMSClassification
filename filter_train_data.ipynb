```python
# libraries

import os
import pandas as pd
```

```python
# files directory
file_dir = "C:/Users/Luke Villanueva/source/repos/LukVill/Big Data - Local/hms-harmful-brain-activity-classification"

# workspace directory
work_dir = "C:/Users/Luke Villanueva/source/repos/LukVill/HMSClassification"
```

```python
# import train csv
#os.path.realpath(file_dir + "/test")
train = pd.read_csv(file_dir + "/train.csv")
train.head()
```

```python
# get unique eeg ids

unique_eeg_id = train["eeg_id"].drop_duplicates().reset_index(drop=True)
unique_eeg_id.head()
```

```python
# parse through eeg folder to find non matching 

# for each eeg id
for id in unique_eeg_id:
    for file in os.listdir(os.path.realpath(file_dir + "/train_eegs")):
        if not file.startswith(str(id)):
            print(file)
            break
```
